# Introduction

> A quick intro to sperm whale vocalizations

Sperm whales are magnificent creatures. They are the largest predator to inhabit planet Earth. They are also the loudest animal - they can produce vocalizations at around 236 db (that is louder than a jet engine)!

They can hear each other from thousands of miles away and some researchers believe that they are able to keep in contact with one another on other sides of the planet.<br>

Their vocalizations are called clicks and this is what they sound like:

```python
from IPython.lib.display import Audio
import librosa

x, rate = librosa.load('data/audio/72009002.wav', sr=None)
Audio(x, rate=rate)

# a recording from the Best Of cuts from the William A. Watkins Collection of Marine Mammal Sound Recordings database
# from Woods Hole Oceanographic Institution (https://cis.whoi.edu/science/B/whalesounds/index.cfm)
```

We can take a look at waveplot to better understand what is going on.

```python
import librosa.display
librosa.display.waveplot(x, rate)
```

This is a plot of water pressure as recorded by the microphone over time. The clicks are the high amplitude vertical lines.

Clicks can form a coda which is a unit of social communication among the whales, though its meaning is yet unclear.

For a great explanation of various forms of vocalizations (with examples!) please consult this amazing [wikipedia page](https://en.wikipedia.org/wiki/Sperm_whale#Vocalization_complex).

The time span between clicks is called the Interclick Interval (ICI). As it turns out, there is a lot of information that can be derived from this piece of data.

Studying the ICI phenomena is the focus of [Deep Machine Learning Techniques for the Detection and Classification of Sperm Whale Bioacoustics](https://www.nature.com/articles/s41598-019-48909-4). In this repository we will reproduce the results and also see if we can extend some of the methods. Ideally, we would also like to propose a novel way of studying the data using Random Forests. That is if we can demonstrate that it is applicable to this kind of problem!
